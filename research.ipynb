### 1. What is gemini-2.5-flash model?

- Introduction: 
    - Gemini 2.5 Flash is a multimodal AI model developed by Google, part of the Gemini 2.5 family released in 2025.

- Key Features:
    - 1. Thinking Capabilities
        - Model with built-in thinking or hybird reasoning.

    - 2. Multimodal Support:
        - Handles text, images, audio, vedio, and code.

    - 3. Context window:
        - Up to 1 million tokens for processing.

    - 4. Strengths:
        - Excels in reasoning, coding, math etc.


- NOTE:
    - gemini-2.5-flash is a large language model developed by google and it's hosting on google Cloud.
